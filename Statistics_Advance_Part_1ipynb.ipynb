13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

Importance:

It allows us to use normal distribution approximations for various statistical analyses.

It is crucial for hypothesis testing and confidence intervals, even when the original data is not normally distributed.


14. How does the Central Limit Theorem relate to the normal distribution?
The Central Limit Theorem (CLT) and the normal distribution are closely related.

....> Relationship between CLT and Normal Distribution:
1. Approximation: The CLT states that the distribution of sample means will be approximately normally distributed, regardless of the underlying distribution of the population, if certain conditions are met.
2. Convergence: As the sample size increases, the distribution of sample means converges to a normal distribution, even if the underlying population distribution is not normal.
3. Normal Distribution as a Limiting Distribution: The normal distribution can be thought of as a limiting distribution, which the distribution of sample means approaches as the sample size increases.

Implications:
1. Universal Applicability: The normal distribution can be used to approximate the distribution of sample means, regardless of the underlying population distribution.
2. Simplification of Statistical Analysis: The normal distribution simplifies statistical analysis, as many statistical tests and procedures assume normality.

21.> What are the assumptions for applying the Central Limit Theorem?

...> The Central Limit Theorem (CLT) is a statistical concept that states that the distribution of sample means will be approximately normally distributed, regardless of the underlying distribution of the population, if certain conditions are met.

Assumptions for Applying the Central Limit Theorem:
1. Independence: Each observation in the sample must be independent of the others. This means that the selection of one observation should not influence the selection of another.
2. Random Sampling: The sample must be selected randomly from the population. This ensures that every member of the population has an equal chance of being selected.
3. Large Sample Size: The sample size (n) must be sufficiently large. A general rule of thumb is that n should be at least 30. However, this can vary depending on the underlying distribution of the population.
4. Finite Population Variance: The population variance (σ^2) must be finite. This ensures that the sample variance is a meaningful estimate of the population variance.
5. Identically Distributed: The observations in the sample must be identically distributed. This means that each observation must come from the same population with the same distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# 1. Generate a random variable and display its value
random_variable = np.random.rand()
print("Random Variable:", random_variable)

```

```python
# 2. Generate a discrete uniform distribution and plot PMF
x = np.arange(1, 7)  # Example dice roll outcomes
pmf = np.ones_like(x) / len(x)
plt.bar(x, pmf)
plt.title("Discrete Uniform Distribution (PMF)")
plt.show()
```

```python

# 3. Probability Distribution Function (PDF) of Bernoulli
p = 0.5
bernoulli_rv = stats.bernoulli(p)
x = [0, 1]
pmf = bernoulli_rv.pmf(x)
plt.bar(x, pmf)
plt.title("Bernoulli Distribution (PDF)")
plt.show()
```

```python
# 4. Simulate binomial distribution (n=10, p=0.5) and plot histogram
binomial_rv = np.random.binomial(n=10, p=0.5, size=1000)
plt.hist(binomial_rv, bins=11, density=True, alpha=0.6, color='g')
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.show()
```

```python
# 5. Create a Poisson distribution and visualize
poisson_rv = np.random.poisson(lam=3, size=1000)
plt.hist(poisson_rv, bins=15, density=True, alpha=0.6, color='b')
plt.title("Poisson Distribution")
plt.show()
```

```python
# 7. Generate and visualize continuous uniform distribution
continuous_uniform = np.random.uniform(0, 10, 1000)
plt.hist(continuous_uniform, bins=30, density=True)
plt.title("Continuous Uniform Distribution")
plt.show()
```

```python

# 18. Calculate Z-score for a given point and compare to standard normal
data_point = 1.5
z_score = (data_point - np.mean(normal_data)) / np.std(normal_data)
print("Z-score for given data point:", z_score)
```

```python
# 19. Implement hypothesis testing using Z-statistics
sample_mean = np.mean(normal_data)
pop_mean = 0
pop_std = 2
z_stat = (sample_mean - pop_mean) / (pop_std / np.sqrt(len(normal_data)))
print("Z-statistic:", z_stat)
```

```python

# 20. Create and interpret confidence interval
confidence = 0.95
z_critical = stats.norm.ppf(1 - (1 - confidence) / 2)
margin_error = z_critical * (pop_std / np.sqrt(len(normal_data)))
conf_interval = (sample_mean - margin_error, sample_mean + margin_error)
print("Confidence Interval:", conf_interval)
```

```python
# 21. Generate normal data and interpret confidence interval for mean
new_data = np.random.normal(50, 10, 1000)
mean_new = np.mean(new_data)
std_new = np.std(new_data, ddof=1)
margin_error_new = z_critical * (std_new / np.sqrt(len(new_data)))
conf_interval_new = (mean_new - margin_error_new, mean_new + margin_error_new)
print("Confidence Interval for New Data Mean:", conf_interval_new)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# 23. Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.
lam = 3
x = np.arange(0, 15)
poisson_cdf = stats.poisson.cdf(x, mu=lam)
plt.step(x, poisson_cdf, where='mid')
plt.title("Poisson Distribution CDF")
plt.xlabel("y")
plt.ylabel("Cumulative Probability")
plt.show()
```

```python
# 25. Write a Python program to compare the standard deviations of two datasets and visualize the difference.
data1 = np.random.normal(50, 10, 1000)  # Dataset 1 (mean=50, std=10)
data2 = np.random.normal(60, 15, 1000)  # Dataset 2 (mean=60, std=15)

std1 = np.std(data1)
std2 = np.std(data2)

plt.hist(data1, bins=40, alpha=0.4, label=f"Dataset 1 (std={std1:.2f})")
plt.hist(data2, bins=40, alpha=0.4, label=f"Dataset 2 (std={std2:.2f})")
plt.legend()
plt.title("Comparison of Standard Deviations")
plt.show()
```

```python

# 26. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.
dataset = np.random.normal(50, 10, 1000)
range_value = np.max(dataset) - np.min(dataset)
iqr = np.percentile(dataset, 70) - np.percentile(dataset, 25)
print(f"Range: {range_value}, IQR: {iqr}")
```

```python
# 27. Implement Z-score normalization on a dataset and visualize its transformation.
z_normalized = (dataset - np.mean(dataset)) / np.std(dataset)
plt.hist(z_normalized, bins=30, density=True)
plt.title("Z-score Normalized Data")
plt.show()
```

```python

# 28. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.
skewness = stats.skew(dataset)
kurtosis = stats.kurtosis(dataset)
print(f"Skewness: {skewness}, Kurtosis: {kurtosis}")
```
